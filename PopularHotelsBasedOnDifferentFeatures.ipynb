```python
import pandas as pd
data = pd.read_csv('Hotel_Reviews.csv')
data.head()
```

```python
# number of distinct features a hotel is known for
print("Total numberof distinct features of hotels : ",len(data.groupby('Tags')['Hotel_Address'].sum()))
```

```python
# list of uniqur tags
df = pd.read_csv("Hotel_Reviews.csv",converters={"Tags": lambda x: x.strip("[]").split(", ")})
values = []
for tags in df['Tags']:
    for i in tags:
        values.append(i.strip("''"))
uniqueTags = set(values)
uniqueTags
```

```python
# https://davidhamann.de/2017/06/26/pandas-select-elements-by-string/
topLeisureHotels = data[data['Tags'].str.contains('Leisure')].groupby('Hotel_Name')['Hotel_Address'].count().sort_values(ascending=False).head(10)
from matplotlib import pyplot as plt
s = pd.Series([1,2,3], index=['a','b','c'])
s.plot.bar(figsize=(20,10))
plt.xlabel('Hotel')
plt.ylabel('Leisure')
plt.title('Top 10 Popular Hotels')
topLeisureHotels.plot(kind = 'bar')
plt.show()
```
