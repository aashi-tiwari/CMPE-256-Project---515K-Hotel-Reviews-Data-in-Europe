```python
import pandas as pd
data = pd.read_csv('Hotel_Reviews.csv')
data.head()
```

```python
#look for any empty value
print(len(data.Hotel_Address) - data.Hotel_Address.count())
print(len(data.Reviewer_Score) - data.Reviewer_Score.count())
#no empty value found
```

```python
topHotels = (data.groupby('Hotel_Name')['Reviewer_Score'].sum()/data.groupby('Hotel_Name')['Reviewer_Score'].count()).sort_values(ascending=False).head(10)
topHotels
```

```python
from matplotlib import pyplot as plt
s = pd.Series([1,2,3], index=['a','b','c'])
s.plot.bar(figsize=(20,10))
plt.xlabel('Hotel')
plt.ylabel('Popularity')
topHotels.plot(kind = 'bar')
plt.title('Top 10 Popular Hotels')
topHotels.plot(kind = 'bar')
plt.show()
```
