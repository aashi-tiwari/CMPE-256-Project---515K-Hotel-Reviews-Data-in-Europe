```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
clean_reviews = pd.read_csv('./data/Hotel_Reviews.csv')
```

```python
#getting top 10 hotel based on reviewer score
topHotelsdf = (clean_reviews.groupby('Hotel_Name')['Reviewer_Score'].sum()/clean_reviews.groupby('Hotel_Name')['Reviewer_Score'].count()).sort_values(ascending=False).head(10)
```

```python
topHotelsdf
```

```python
TopHotels = ['Ritz Paris',                     
'Hotel Casa Camper',                     
'41',                                   
'H tel de La Tamise Esprit de France',   
'Le Narcisse Blanc Spa','H10 Casa Mimosa 4 Sup',
'Hotel Eiffel Blomet',
'Hotel The Serras' ,
'45 Park Lane Dorchester Collection',
'The Soho Hotel']
```

```python
#plotting graph between date user reviewed and rating given by user for top 10 hotels.
for h in TopHotels:
    Hotel1 = clean_reviews[clean_reviews.Hotel_Name == h]
    Hotel1.plot('Review_Date', 'Reviewer_Score',figsize=(10,5))
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Review_score', fontsize=12)
    title_str = h
    plt.title(title_str, fontsize=15)
    plt.show()
```
